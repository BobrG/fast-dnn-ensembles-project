```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'seed_303_seed_42_mse'
curve_mse = np.load(f'./curves_data/{filename}/curve.npz')
```

```python
def plot_curve(curve, name='', loss_name=''):
    fig, ax = plt.subplots(1, 2, figsize=(25, 10))
    
    ax[0].scatter([0.0], [curve['tr_loss'][0]], c='black', label='w0')
    ax[0].scatter([1.0], [curve['tr_loss'][-1]], c='black', label='w1')
    ax[0].plot(curve['ts'], curve['tr_loss'], label=f'loss {loss_name}')
    ax[0].set_ylabel('Loss Train', fontsize=16)
    ax[0].set_xlabel('t', fontsize=16)
    ax[0].grid('on')
    ax[0].legend(fontsize=16)
    ax[0].set_ylim(top=max(curve['tr_loss'])+0.001)
    ax[0].set_ylim(bottom=min(curve['tr_loss'])-0.001)
    
    ax[1].scatter([0.0], [curve['te_loss'][0]], c='black', label='w0')
    ax[1].scatter([1.0], [curve['te_loss'][-1]], c='black', label='w1')
    ax[1].plot(curve['ts'], curve['te_loss'], label=f'loss {loss_name}')
    ax[1].set_ylabel('Loss Train', fontsize=16)
    ax[1].set_xlabel('t', fontsize=16)
    ax[1].grid('on')
    ax[1].legend(fontsize=16)
    ax[1].set_ylim(top=max(curve['te_loss'])+0.001)
    ax[1].set_ylim(bottom=min(curve['te_loss'])-0.001)
    
    fig.suptitle(f'Curve 30 points {name}')
    
    
```

```python
plot_curve(curve_mse, name=filename, loss_name='mse')
```

```python
filename = 'seed_303_seed_42_mse'
line_mse_303_42 = np.load(f'./segment/{filename}/chain.npz', allow_pickle=True)
```

```python
plot_curve(line_mse_303_42, name=filename, loss_name='mse')
```

```python
plot_curve(curve_vgg, name=filename+' CurveNet', loss_name='vgg')
```

```python
filename = 'seed_303_seed_42_vgg'
segment_vgg = np.load(f'./segment/{filename}/chain.npz')
```

```python
plot_curve(segment_vgg, name=filename+' Segment', loss_name='vgg')
```

```python
def plot_curve_and_segment(curve, segment, name='', loss_name=''):
    fig, ax = plt.subplots(1, 2, figsize=(25, 10))
    
    ax[0].scatter([0.0], [curve['tr_loss'][0]], c='black', label='w0')
    ax[0].scatter([1.0], [curve['tr_loss'][-1]], c='black', label='w1')
    ax[0].plot(curve['ts'], curve['tr_loss'], label=f'Curve loss {loss_name}')
    ax[0].plot(segment['ts'], segment['tr_loss'], label=f'Segment loss {loss_name}')
    ax[0].set_ylabel('Loss Train', fontsize=16)
    ax[0].set_xlabel('t', fontsize=16)
    ax[0].grid('on')
    ax[0].legend(fontsize=16)
    ax[0].set_ylim(top=max(segment['tr_loss'])+0.001)
    ax[0].set_ylim(bottom=min(curve['tr_loss'])-0.01)
    
    ax[1].scatter([0.0], [curve['te_loss'][0]], c='black', label='w0')
    ax[1].scatter([1.0], [curve['te_loss'][-1]], c='black', label='w1')
    ax[1].plot(curve['ts'], curve['te_loss'], label=f'loss {loss_name}')
    ax[1].plot(segment['ts'], segment['te_loss'], label=f'Segment loss {loss_name}')
    ax[1].set_ylabel('Loss Test', fontsize=16)
    ax[1].set_xlabel('t', fontsize=16)
    ax[1].grid('on')
    ax[1].legend(fontsize=16)
    ax[1].set_ylim(top=max(segment['te_loss'])+0.001)
    ax[1].set_ylim(bottom=min(curve['te_loss'])-0.01)
    
    fig.suptitle(f'Curve 30 points {name}')
    
    
```

```python
plot_curve_and_segment(curve_vgg, segment_vgg, name='seed_303_seed_42', loss_name='vgg')
```
